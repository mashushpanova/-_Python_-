# 1. Парадокс инспектора

### Интервалы 5\10 минут 

```python
# Генерация данных

res = []
bus_data = []
for i in range(10** 4):
    day = (np.random.binomial(1.0, 0.5, 144) + 1) * 5
    while sum(day) > 720:
        day = np.delete(day, -1)
        
    res.append(len(day))
    bus_data += [day]
```

```python
# Гистограмма числа автобусов
# Визуально распределение похоже на нормальное

unique, counts = np.unique(res, return_counts = True)
d = dict(zip(unique, counts))

plt.figure(figsize = (15, 7))
plt.bar(d.keys(), d.values())
plt.title("Гистограмма количества автобусов за сутки")
plt.xlabel("Количество автобусов за сутки")
plt.ylabel("Сколько дней было данное количество")
plt.show()
```

```python
# Инспекция инспектора Тимона

timon = list(map(lambda x: np.random.choice(x), bus_data))
unique, counts = np.unique(timon, return_counts = True)
d = dict(zip(unique, counts))

print("Оценка матожидания: {}".format(np.mean(timon)))
print("Оценка дисперсии: {}".format(np.var(timon)))

plt.bar(["5", "10"], d.values())
plt.title("Инспекция Тимона")
plt.xlabel("Количество пассажиров в автобусе")
plt.ylabel("Сколько дней было данное количество")
plt.show()
```

```python
# Инспекция инспектора Пумбы

pumba = []
for day in bus_data:
    pumba_time = np.random.uniform(0, 721)
    time = 0.0
    i = 0
    while time < pumba_time and i < len(day):
        time += day[i]
        i += 1
    
    if i < len(day):
        pumba.append(day[i - 1])
    
unique, counts = np.unique(pumba, return_counts = True)
d = dict(zip(unique, counts))

print("Оценка матожидания: {}".format(np.mean(pumba)))
print("Оценка дисперсии: {}".format(np.var(pumba)))

plt.bar(["5", "10"], d.values())
plt.title("Инспекция Пумбы")
plt.xlabel("Количество пассажиров в автобусе")
plt.ylabel("Сколько дней было данное количество")
plt.show()
```

# 2. Парадокс Паррондо

```python
atos_result = []
portos_result = []
aramis_result = []


for n in range(10 ** 4):
    # Атос
    atos_result.append([])
    atos = np.random.choice([1, -1], p = [0.49, 0.51], size = 1000)
    atos_result[n].append(1000)
    atos_money = 1000
    for i in atos:
        atos_money += i
        atos_result[n].append(atos_money)
    

    # Портос
    portos_result.append([])
    portos_result[n].append(1000)
    portos_money = 1000
    for i in range(1000):
        if portos_money % 3 == 0:
            tmp = np.random.choice([1, -1], p = [0.09, 0.91])
            portos_money += tmp
            portos_result[n].append(portos_money)
        else:
            tmp = np.random.choice([1, -1], p = [0.74, 0.26])
            portos_money += tmp
            portos_result[n].append(portos_money)
        
        
    # Арамис
    aramis_result.append([])
    aramis_result[n].append(1000)
    aramis_money = 1000
    for i in range(1000):
        if np.random.randint(0, 2) == 0:
            if aramis_money % 3 == 0:
                tmp = np.random.choice([1, -1], p = [0.09, 0.91])
                aramis_money += tmp
                aramis_result[n].append(aramis_money)
            else:
                tmp = np.random.choice([1, -1], p = [0.74, 0.26])
                aramis_money += tmp
                aramis_result[n].append(aramis_money)
        else:
            tmp = np.random.choice([1, -1], p = [0.49, 0.51])
            aramis_money += tmp
            aramis_result[n].append(aramis_money)
```

```python
# Атос
plt.figure(figsize = (15, 7))
plt.plot(list(range(1001)), np.sum(atos_result, axis = 0) / len(atos_result), label = "Атос среднее")
for i in range(10):
    plt.plot(list(range(1001)), atos_result[np.random.randint(0, len(atos_result))], label = "Атос {}".format(i))
plt.title("Богатство Атоса")
plt.xlabel("Номер посещения казино")
plt.ylabel("Богатство")
plt.legend()
plt.show()
    
```

```python
# Портос
plt.figure(figsize = (15, 7))
plt.plot(list(range(1001)), np.sum(portos_result, axis = 0) / len(portos_result), label = "Портос среднее")
for i in range(10):
    plt.plot(list(range(1001)), portos_result[np.random.randint(0, len(portos_result))], label = "Портос {}".format(i))
plt.title("Богатство Портоса")
plt.xlabel("Номер посещения казино")
plt.ylabel("Богатство")
plt.legend()
plt.show()
```

```python
# Арамис 
plt.figure(figsize = (15, 7))
plt.plot(list(range(1001)), np.sum(aramis_result, axis = 0) / len(aramis_result), label = "Арамис среднее")
for i in range(10):
    plt.plot(list(range(1001)), aramis_result[np.random.randint(0, len(aramis_result))], label = "Арамис {}".format(i))
plt.title("Богатство Арамиса")
plt.xlabel("Номер посещения казино")
plt.ylabel("Богатство")
plt.legend()
plt.show()
```

# 3. Парадокс Берксона

```python
students = []
corr_students = []
corr_no_students = []

for n in range(80, 161, 5):
    math = np.random.normal(60, 10, 10 ** 4)
    russian = np.random.normal(60, 10, 10 ** 4)
    
    res = math + russian
    students.append(len(res[res > n]))
    
    corr_students.append(np.correlate(math[math + russian > n], russian[math + russian > n]))
    
    corr_no_students.append(np.correlate(math[math + russian <= n], russian[math + russian <= n]))
    
plt.figure(figsize = (10, 5))
plt.plot(list(range(80, 161, 5)), students, label = "Прошедшие")
plt.title("Количество прошедших в зависимости от проходного балла n")
plt.xlabel("Проходной балл n")
plt.ylabel("Количество прошедших")
plt.legend()
plt.show()

plt.figure(figsize = (10, 5))
plt.plot(list(range(80, 161, 5)), corr_students, label = "Корреляция среди прошедших", color = 'g')
plt.plot(list(range(80, 161, 5)), corr_no_students, label = "Корреляция среди не прошедших", color = 'r')
plt.title("Корреляция между результатами по русскому и математике")
plt.xlabel("Проходной балл n")
plt.ylabel("Коэффициент корреляции")
plt.legend()
plt.show()
```

```python
# Волк знает дисперсию = 100
# Волк знает значение сл.вел. на каждый день
# Оценка максимального правдоподобия для матожидания 
# нормального распределения - выборочное среднее

mx_all = np.array([])
my_all = np.array([])
mz_all = np.array([])

for i in range(10 ** 4):
    mx_all = np.append(mx_all, np.mean(np.array(np.random.normal(60, 100, 100), dtype = np.int32)))
    my_all = np.append(my_all, np.mean(np.array(np.random.normal(70, 100, 100), dtype = np.int32)))
    mz_all = np.append(mz_all, np.mean(np.array(np.random.normal(80, 100, 100), dtype = np.int32)))
```

```python
d1 = dict(zip(*np.unique(mx_all, return_counts = True)))
d2 = dict(zip(*np.unique(my_all, return_counts = True)))
d3 = dict(zip(*np.unique(mz_all, return_counts = True)))

plt.figure(figsize = (15, 7))
plt.bar(d1.keys(), d1.values(), label = 'Оценка Х')
plt.bar(60, max( d1.values()), width = 0.3, color = 'r')
plt.bar(d2.keys(), d2.values(), label = 'Оценка У')
plt.bar(70, max( d2.values()), width = 0.3, color = 'r')
plt.bar(d3.keys(), d3.values(), label = 'Оценка Z')
plt.bar(80, max( d3.values()), width = 0.3, color = 'r')
plt.title("Гистограмма оценок")
plt.xlabel("Значение оценки")
plt.ylabel("Количество")
plt.legend()
plt.show()
```

```python
S_volk = (mx_all - 60) ** 2 + (my_all - 70) ** 2 + (mz_all - 80) ** 2
print("Оценка матожидания - {}".format(np.mean(S_volk)))

plt.figure(figsize = (15, 7))
d = dict(zip(*np.unique(np.array(S_volk, dtype = np.int32), return_counts = True)))
plt.bar(d.keys(), d.values())
plt.title("Гистограмма суммарной квадратичной ошибки")
plt.xlabel("Значение ошибки")
plt.ylabel("Количество")
plt.show()
```
